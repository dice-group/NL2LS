```python
import os
import time
from openai import OpenAI
import pandas as pd
import time

client = OpenAI(
    # This is the default and can be omitted
    api_key="<Your-OpenAI-Key",
)

def query_llm(prompt):
    """
    Send a prompt to the LLM and return the response.
    """
    response = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        model="gpt-3.5-turbo-0125",
    )
    return response.choices[0].message.content

def main():
    # Define the domain and scope
    domain = "Link specifications"
    scope = "Generate link specification from natural language"
    dataset = "Link specifications"
    model="gpt-3.5"
    # Read CSV files
    df = pd.read_csv(f"../../datasets/New-Datasets/fine-tuning/silk-human-annotated/test.txt", sep="\t")
    test_list = df['1'].tolist()
    ls_list = df['0'].tolist()
    f = open(f"results.txt", "w")
    f_ls = open(f"ls.txt", "w")
    for num, nl in enumerate(test_list):
        print(f"generating ls {num+1}/{len(test_list)}")
        prompt = f"""
            Translate text into Link Specification: 
            {nl}
            """
        output = query_llm(prompt)
        print(output)
        f.write(f"{output}\n")
        f_ls.write(f"{ls_list[num]}\n")
        time.sleep(10)
    f.close()
    f_ls.close()
if __name__ == "__main__":
    main()
```
